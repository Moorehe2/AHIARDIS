```python
heartdata = pd.read_csv('https://raw.githubusercontent.com/hantswilliams/AHI_DataSci_507/main/Datasets/HeartDisease/heart.csv')
heartdata
```

```python
df = heartdata[['trestbps', 'chol']]
df = df.rename({'trestbps': 'systolicBP', 'chol': 'cholesterol'}, axis='columns')
df
```

```python
# Load in linear regression 
lin_reg = LinearRegression()
```

```python
## Non-ARRAY 
df['systolicBP']
```

```python
## ARRAY 
df[['systolicBP']]
```

```python
dv = df[['systolicBP']]
iv = df[['cholesterol']]
```

```python
# Load in variables of interest into linear regression 
## x = systolicBP 
## y = cholesterol 

lin_reg.fit(dv, iv)
```

```python
# Print out statistical markers 
print(f'Coefficients: {lin_reg.coef_}')
print(f'Intercept: {lin_reg.intercept_}')
print(f'R^2 score: {lin_reg.score(dv, iv)}')
```
